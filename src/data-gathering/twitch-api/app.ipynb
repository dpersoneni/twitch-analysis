```python
from twitchAPI.twitch import Twitch

# create instance of twitch API
twitch = Twitch('client-id', 'client-secret')
twitch.authenticate_app([])
```

```python
streamers = []
channels = open("channels.txt", "r").readlines()
for channel in channels:
    if "\n" in channel:
        channel = channel.rstrip("\n")
  
    streamers.append(channel)
    
```

```python
for streamer in streamers:
    if ("[" in streamer or "]" in streamer):
        print(streamer)
```

```python
f = open("graph.txt", "w+")
for streamer in streamers:
    streamer_info = twitch.get_users(logins=['{}'.format(streamer)])
    if (streamer_info['data'] != [] ): 
        streamer_id = streamer_info['data'][0]['id']

        followers = twitch.get_users_follows(from_id=streamer_id, first=100)
        total_requests = followers['total'] // 100

        people = followers['data']
        
        for person in people:
            f.write("%s %s\n" % (streamer, person['to_login']))

        if (followers['pagination'] != {} ): 
            pagination = followers['pagination']['cursor']
            i = 0
            while i < total_requests:
                followers = twitch.get_users_follows(from_id=streamer_id, first=100, after=pagination)
                people = people + followers['data']
                for person in people:
                    f.write("%s %s\n" % (streamer, person['to_login']))


                if (followers['pagination'] == {} ): 
                    print(i)
                    break
                pagination = followers['pagination']['cursor']
                i += 1

        
f.close()

# for id in followers['data']:
```
